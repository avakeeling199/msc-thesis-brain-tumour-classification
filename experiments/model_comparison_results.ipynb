```python
# load in data
import pandas as pd

all_results = pd.read_csv("../results/all_model_results.csv")
```

```python
all_results
```

```python
summary = all_results.groupby(["Architecture", "Classifier"]).agg({
    "Accuracy": ["mean", "std"],
    "F1 Score": ["mean", "std"],
    "Precision": ["mean", "std"],
    "Recall": ["mean", "std"],
    "Specificity": ["mean", "std"],
    "ROC AUC": ["mean", "std"],
    "training_time": "mean",
}).reset_index()
```

```python
summary
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(
    data=summary,
    x="Architecture",
    y=("Accuracy", "mean"),
    hue="Classifier",
    capsize=0.1
)
plt.title("Mean Accuracy by Architecture and Classifier")
plt.ylabel("Accuracy")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(
    data=summary,
    x="Architecture",
    y=("F1 Score", "mean"),
    hue="Classifier",
    capsize=0.1
)
plt.title("Mean F1 Score by Architecture and Classifier")
plt.ylabel("Accuracy")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_f1 = summary.sort_values(by=("F1 Score", "mean"), ascending=False).head(5)
print(top_f1)
```

```python
top_auc = summary.sort_values(by=("ROC AUC", "mean"), ascending=False).head(5)
print(top_auc)
```

```python
top_acc = summary.sort_values(by=("Accuracy", "mean"), ascending=False).head(5)
print(top_acc)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=all_results,
    x="Architecture",
    y="training_time",
    hue="Classifier"
)
plt.title("Training Time by Architecture and Classifier (10 folds)")
plt.ylabel("Time (seconds)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
